```python
#===========================================================
# morphology
#===========================================================
#---------------------------------------
# Erosion and Dilation
#---------------------------------------
import numpy as np
import cv2
import os
from matplotlib import pyplot as plt
Original = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Lesson04_IP.rar (Unzipped Files)/Images_New/opencv.png', 0)
Original.shape

#Original.dtype
kernel = np.ones((5, 5), np.uint8) # structure element
# erosion
erosion = cv2.erode(Original, kernel, iterations = 1)
# dilation
dilation = cv2.dilate(Original, kernel, iterations = 1)

plt.subplot(311)
plt.xticks([])
plt.yticks([])
plt.imshow(Original,plt.cm.gray)
plt.subplot(312)
plt.xticks([])
plt.yticks([])
plt.imshow(erosion,plt.cm.gray)
plt.subplot(313)
plt.xticks([])
plt.yticks([])
plt.imshow(dilation,plt.cm.gray)
plt.show()
```

```python
#---------------------------------------
# Open and Close
#---------------------------------------
test = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Lesson04_IP.rar (Unzipped Files)/Images_New/test.jpg', 0)
test.shape
plt.xticks([])
plt.yticks([])
plt.imshow(test,plt.cm.gray)
plt.show()

kernel = np.ones((5, 5), np.uint8)

#-----open=erosion+dilation
erosion = cv2.erode(test, kernel, iterations = 1)
result_opening = cv2.dilate(erosion, kernel, iterations = 1)
plt.xticks([])
plt.yticks([])
plt.xlabel('opening')
plt.imshow(result_opening,plt.cm.gray)
plt.show()

#-----close=dilation+erosion
dilation = cv2.dilate(test, kernel, iterations = 1)
result_closeing = cv2.erode(dilation, kernel, iterations = 1)
plt.xticks([])
plt.yticks([])
plt.xlabel('closeing')
plt.imshow(result_closeing,plt.cm.gray)
plt.show()

plt.xticks([])
plt.yticks([])
plt.imshow(test,plt.cm.gray)
plt.show()
#-----open=erosion+dilation, single routine
result_opening = cv2.morphologyEx(test, cv2.MORPH_OPEN, kernel)
plt.xticks([])
plt.yticks([])
plt.xlabel('opening')
plt.imshow(result_opening,plt.cm.gray)
plt.show()

#-----close=dilation+erosion, single routine
result_closeing = cv2.morphologyEx(test, cv2.MORPH_CLOSE, kernel)
plt.xticks([])
plt.yticks([])
plt.xlabel('closeing')
plt.imshow(result_closeing,plt.cm.gray)
plt.show()
#===========================================================
```
