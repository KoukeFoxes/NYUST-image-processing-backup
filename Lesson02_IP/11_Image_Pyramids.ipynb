```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
A = cv2.imread('Images01/apple.jpg')
B = cv2.imread('Images01/orange.jpg')

# generate Gaussian pyramid for A  
G = A.copy()  
gpA = [G]  
for i in np.arange(6):     #將蘋果進行高斯金字塔處理，總共六級處理  
    G = cv2.pyrDown(G)  
    gpA.append(G)  

# generate Gaussian pyramid for B  
G = B.copy()  
gpB = [G]  
for i in np.arange(6):         #將橘子進行高斯金字塔處理，總共六級處理  
    G = cv2.pyrDown(G)  
    gpB.append(G)  

# generate Laplacian Pyramid for A  
lpA = [gpA[5]]                 
for i in np.arange(5,0,-1):    #將蘋果進行拉普拉斯金字塔處理，總共5級處理 
    GE = cv2.pyrUp(gpA[i])  
    L = cv2.subtract(gpA[i-1],GE)  
    lpA.append(L)  
    
# generate Laplacian Pyramid for B  
lpB = [gpB[5]]  
for i in np.arange(5,0,-1):    #將橘子進行拉普拉斯金字塔處理，總共5級處理   
    GE = cv2.pyrUp(gpB[i])  
    L = cv2.subtract(gpB[i-1],GE)  
    lpB.append(L)  

# Now add left and right halves of images in each level  
# numpy.hstack(tup)  
# Take a sequence of arrays and stack them horizontally  
#  make a single array.  

LS = []  
for la,lb in zip(lpA,lpB):  
    rows,cols,dpt = la.shape  
    #將兩張影像的矩陣的左半部分和右半部分拼接到一起  
    ls = np.hstack((la[:,0:cols//2], lb[:,cols//2:]))   
    LS.append(ls)  

# now reconstruct  
ls_ = LS[0]   #這裡 LS[0] 為高斯金字塔的最小影像 

#第一次循環的影像為高斯金字塔的最小影像，依次通過拉普拉斯金字塔恢覆到大影像 
for i in range(1,6):                         
    ls_ = cv2.pyrUp(ls_)  
    ls_ = cv2.add(ls_, LS[i])                #採用金字塔拼接方法影像  

# image with direct connecting each half  
real = np.hstack((A[:,:cols//2],B[:,cols//2:]))   #直接的拼接  
cv2.imwrite('Images01/Pyramid_blending2.jpg',ls_)  
cv2.imwrite('Images01/Direct_blending.jpg',real)
```
