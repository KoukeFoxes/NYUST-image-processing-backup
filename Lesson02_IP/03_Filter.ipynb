```python
#---------------------------------------
# required packages
#---------------------------------------
import cv2
import numpy as np
from matplotlib import pyplot as plt
#---------------------------------------
# Basic filtering
#---------------------------------------
fn_in="Images01/lena.tiff"
img_in = cv2.imread(fn_in, 0) # 0: convert to gray, 1: color
print(type(img_in))
img_in.shape
img_in.dtype
img_in.nbytes
plt.imshow(img_in,plt.cm.gray)
plt.xticks([])
plt.yticks([])
plt.xlabel('img_in')
plt.show()
```

```python
#-----Average filter
Averaging3 = cv2.blur(img_in,(3, 3))  # Averaging
Averaging3.shape
Averaging3.dtype
Averaging3.nbytes
plt.imshow(Averaging3,plt.cm.gray)
plt.xticks([])
plt.yticks([])
plt.xlabel('Averaging3')
plt.show()

Averaging5 = cv2.blur(img_in,(5, 5))  # Averaging
plt.imshow(Averaging5,plt.cm.gray)
plt.xticks([])
plt.yticks([])
plt.xlabel('Averaging5')
plt.show()

Averaging7 = cv2.blur(img_in,(7, 7))  # Averaging
plt.imshow(Averaging7,plt.cm.gray)
plt.xticks([])
plt.yticks([])
plt.xlabel('Averaging7')
plt.show()


Averaging9 = cv2.blur(img_in,(9, 9))  # Averaging
plt.imshow(Averaging9,plt.cm.gray)
plt.xticks([])
plt.yticks([])
plt.xlabel('Averaging9')
plt.show()
```

```python
#-----Gaussian filter
Gaussian = cv2.GaussianBlur(img_in,(5, 5), 0) # Gaussian Filtering
plt.imshow(Gaussian,plt.cm.gray)
plt.xticks([])
plt.yticks([])
plt.xlabel('Gaussian')
plt.show()

#-----Median filter
fn_in="/content/drive/MyDrive/Colab Notebooks/Lesson04_IP.rar (Unzipped Files)/Images_New/lena_sp_0.05.tiff"

img_in = cv2.imread(fn_in, 0) # 0: convert to gray, 1: color
type(img_in)
img_in.shape
img_in.dtype
img_in.nbytes
plt.imshow(img_in,plt.cm.gray)
plt.xticks([])
plt.yticks([])
plt.xlabel('img_in')
plt.show()

Median = cv2.medianBlur(img_in,3)  # Median Filtering
plt.imshow(Median,plt.cm.gray)
plt.xticks([])
plt.yticks([])
plt.xlabel('Median')
plt.show()
#===========================================================
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
def gaussian_kernel(size, size_y=None):
    size = int(size)
    if not size_y:
        size_y = size
    else:
        size_y = int(size_y)
    x, y = np.mgrid[-size:size+1, -size_y:size_y+1]
    g = np.exp(-(x**2/float(size)+y**2/float(size_y)))
    return g / g.sum()
# Make the Gaussian by calling the function
gaussian_kernel_array = gaussian_kernel(5)
 
plt.imshow(gaussian_kernel_array, cmap=plt.get_cmap('jet'), interpolation='nearest')
plt.colorbar()
plt.show()
```
