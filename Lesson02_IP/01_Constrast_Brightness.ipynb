```python
#---------------------------------------
# simple ndarray operation, neagtion
#---------------------------------------
fn_in="Images01/lena.tiff"
#fn_in="baboon.bmp"
print(fn_in.split("/")[-1])

img_in = cv2.imread(fn_in, 0) # 0: convert to gray, 1: color
type(img_in)
img_in.shape
img_in.dtype
plt.imshow(img_in,plt.cm.gray)
plt.show()

img_out = abs(255-img_in)  #負片
type(img_out)
img_out.shape
img_out.dtype
plt.imshow(img_out,plt.cm.gray)
plt.show()

# save the result image
fn_in = fn_in.split("/")[-1]
fn_tmp=fn_in.split(".")
fn_tmp
type(fn_tmp) # the list
fn_tmp[0]
print(fn_tmp[1])

fn_out=fn_tmp[0]+"_neg."+fn_tmp[1]
cv2.imwrite(fn_out,img_out)
```

```python
#---------------------------------------
# Brightness and contrast adjustment
#---------------------------------------
fn_in="Images01/lena.tiff"
#fn_in="baboon.bmp"

img_in = cv2.imread(fn_in, 0) # 0: convert to gray, 1: color
type(img_in)
img_in.shape
img_in.dtype
plt.imshow(img_in,plt.cm.gray)
plt.show()

#-----brighter 01
img_p = np.int32(img_in)+50
img_out = np.uint8(img_p);
img_out[img_p < 0] = 0
img_out[img_p > 255] = 255
type(img_out)
img_out.shape
img_out.dtype
plt.imshow(img_out,plt.cm.gray)
plt.show()

#-----brighter 02
img = img_in+50.
img[np.where(img>255)]=255
img[np.where(img<0  )]=0
img_out = img.astype("uint8") # truncate directly
plt.imshow(img_out,plt.cm.gray)
plt.show()

#-----darker
img_p = np.int32(img_in)-50
img_out = np.uint8(img_p);
img_out[img_p < 0] = 0
img_out[img_p > 255] = 255
type(img_out)
img_out.shape
img_out
plt.imshow(img_out,plt.cm.gray)
plt.show()
```

```python
#---------------------------------------
# Contrast adjustment
#---------------------------------------
fn_in="Images01/lena.tiff"
#fn_in="baboon.bmp"

img_in = cv2.imread(fn_in, 0) # 0: convert to gray, 1: color
type(img_in)
img_in.shape
img_in.dtype
img_in.nbytes
plt.imshow(img_in,plt.cm.gray)
plt.show()

#-----higher contrast
img_out = img_in*1.2
type(img_out)
img_out.shape
img_out.dtype
img_out=img_out.astype('uint8') # change dtype to unsigned int
plt.imshow(img_out,plt.cm.gray)
plt.show()

img = img_in*1.2
img[np.where(img>255)]=255
img[np.where(img<0  )]=0
img_out=img_out.astype('uint8') # change dtype to unsigned int
img_out.dtype
plt.imshow(img_out,plt.cm.gray)
plt.show()
```

```python
#-----lower contrast
img_out = img_in*0.4
type(img_out)
img_out.shape
img_out.dtype
plt.imshow(img_out,plt.cm.gray)
plt.show()
img_out=img_out.astype('uint8') # change dtype to unsigned int
img_out.dtype
plt.imshow(img_out,plt.cm.gray)
plt.show()

img_out = img_in*0.4
type(img_out)
plt.imshow(img_out,plt.cm.gray)
plt.show()
img_out=img_out.astype('uint8') # change dtype to unsigned int
plt.imshow(img_out,plt.cm.gray)
plt.show()
```
