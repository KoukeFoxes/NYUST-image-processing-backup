```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt
img = cv2.imread('Images01/messi5.jpg',0)
rows,cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])
print(M)
dst = cv2.warpAffine(img,M,(cols,rows))

plt.imshow(dst,plt.cm.gray)
plt.show()
```

```python
#---------------------------------------
# required packages
#---------------------------------------
#pip install image
import numpy as np
import cv2
import os
from matplotlib import pyplot as plt
fname = "Images01/lena.tiff" #file name

#---------------------------------------
# rotation
#---------------------------------------
img = cv2.imread(fname, 0)
ori_img = img
rows, cols = img.shape
rotation = 135
if (rotation%90==0) :
    resize=1
else :
    resize=0.7

#cv2. getRotationMatrix2D (center, angle, scale) 
M = cv2.getRotationMatrix2D((cols/2,rows/2), rotation, resize)
print(M)
dst = cv2.warpAffine(img, M, (rows, cols))

#---------------------------------------
# display an image
#---------------------------------------
plt.imshow(ori_img,plt.cm.gray)
plt.show()
plt.imshow(dst,plt.cm.gray)
plt.show()
```
