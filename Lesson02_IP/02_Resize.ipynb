```python
#---------------------------------------
# required packages
#---------------------------------------
import cv2
import numpy as np
from matplotlib import pyplot as plt
#===========================================================
# resampling
#===========================================================
img = cv2.imread('Images01/bird2.jpg',1)
img.shape
img.dtype

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#-----set scaling factor
resize1 = cv2.resize(img,None,fx=2, fy=2,interpolation=cv2.INTER_LINEAR)
resize1.shape
plt.imshow(cv2.cvtColor(resize1, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
resize1 = cv2.resize(img,None,fx=0.3,fy=0.3,interpolation=cv2.INTER_LINEAR)
resize1.shape
plt.imshow(cv2.cvtColor(resize1, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
resize1 = cv2.resize(img,None,fx=0.6,fy=1.2,interpolation=cv2.INTER_LINEAR)
resize1.shape
plt.imshow(cv2.cvtColor(resize1, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#-----various interpolation methods (default: cv2.INTER_LINEAR)
#img = cv2.imread('rilakkuma.jpg',1)
img = cv2.imread('Images01/lena.tiff',1)
print(img.shape)
```

```python
img = cv2.resize(img,None,fx=.5, fy=.5,interpolation=cv2.INTER_LINEAR)
print(img.shape)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
r_NEAREST = cv2.resize(img,None,fx=5, fy=5,interpolation=cv2.INTER_NEAREST)
plt.imshow(cv2.cvtColor(r_NEAREST, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
r_LINEAR  = cv2.resize(img,None,fx=5, fy=5,interpolation=cv2.INTER_LINEAR)
plt.imshow(cv2.cvtColor(r_LINEAR, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
r_CUBIC   = cv2.resize(img,None,fx=5, fy=5,interpolation=cv2.INTER_CUBIC)
plt.imshow(cv2.cvtColor(r_CUBIC, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
r_LANCZOS4= cv2.resize(img,None,fx=5, fy=5,interpolation=cv2.INTER_LANCZOS4)
plt.imshow(cv2.cvtColor(r_LANCZOS4, cv2.COLOR_BGR2RGB))
plt.show()
```
